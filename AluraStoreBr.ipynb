### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
# @title Preço

from matplotlib import pyplot as plt
loja['Preço'].plot(kind='hist', bins=20, title='Preço')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

#1. Análise do faturamento


```python
total_faturamento_loja1 = loja['Preço'].sum()
total_faturamento_loja2 = loja2['Preço'].sum()
total_faturamento_loja3 = loja3['Preço'].sum()
total_faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento da Loja 1: R${total_faturamento_loja1:,.2f}")
print(f"Faturamento da Loja 2: R${total_faturamento_loja2:,.2f}")
print(f"Faturamento da Loja 3: R${total_faturamento_loja3:,.2f}")
print(f"Faturamento da Loja 4: R${total_faturamento_loja4:,.2f}")

faturamento_total = total_faturamento_loja1 + total_faturamento_loja2 + total_faturamento_loja3 + total_faturamento_loja4
print(f"\nFaturamento Total das Lojas: R${faturamento_total:,.2f}")


faturamento_por_loja = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento': [total_faturamento_loja1, total_faturamento_loja2, total_faturamento_loja3, total_faturamento_loja4]
})

##Gráfico de fatauramento das Lojas
faturamento_por_loja.plot(kind='bar', x='Loja', y='Faturamento', legend=False, title='Faturamento por Loja')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=0)
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

# 2. Vendas por Categoria


```python
vendas_categoria_loja1 = loja.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
vendas_categoria_loja2 = loja2.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
vendas_categoria_loja3 = loja3.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
vendas_categoria_loja4 = loja4.groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)

print("Vendas por Categoria - Loja 1:")
print(vendas_categoria_loja1)
print("\nVendas por Categoria - Loja 2:")
print(vendas_categoria_loja2)
print("\nVendas por Categoria - Loja 3:")
print(vendas_categoria_loja3)
print("\nVendas por Categoria - Loja 4:")
print(vendas_categoria_loja4)

# Total de vendas por categoria em todas as lojas
vendas_total_categoria = pd.concat([loja, loja2, loja3, loja4]).groupby('Categoria do Produto')['Preço'].sum().sort_values(ascending=False)
print("\nTotal de Vendas por Categoria (Todas as Lojas):")
print(vendas_total_categoria)

# Vendas por loja e categoria
vendas_por_loja_categoria = pd.concat([loja.assign(Loja='Loja 1'),
                                       loja2.assign(Loja='Loja 2'),
                                       loja3.assign(Loja='Loja 3'),
                                       loja4.assign(Loja='Loja 4')]).groupby(['Loja', 'Categoria do Produto'])['Preço'].sum()
print("\nVendas por Loja e Categoria:")
print(vendas_por_loja_categoria)

# Opcional: Visualizar as vendas por categoria para cada loja em gráficos separados
vendas_categoria_loja1.plot(kind='line', title='Vendas por Categoria - Loja 1')
plt.ylabel('Vendas (R$)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

vendas_categoria_loja2.plot(kind='line', title='Vendas por Categoria - Loja 2')
plt.ylabel('Vendas (R$)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

vendas_categoria_loja3.plot(kind='line', title='Vendas por Categoria - Loja 3')
plt.ylabel('Vendas (R$)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

vendas_categoria_loja4.plot(kind='line', title='Vendas por Categoria - Loja 4')
plt.ylabel('Vendas (R$)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Gráfico do total de vendas por categoria em todas as lojas
vendas_total_categoria.plot(kind='line', title='Total de Vendas por Categoria (Todas as Lojas)')
plt.ylabel('Vendas (R$)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Unindo os dados das quatro lojas com uma coluna identificando a loja
df_total = pd.concat([loja.assign(Loja='Loja 1'),
                      loja2.assign(Loja='Loja 2'),
                      loja3.assign(Loja='Loja 3'),
                      loja4.assign(Loja='Loja 4')])

# Agrupando por Categoria do Produto e Loja e somando o preço (vendas)
vendas_por_categoria_loja_pivot = df_total.groupby(['Categoria do Produto', 'Loja'])['Preço'].sum().unstack()

# Preenchendo possíveis valores NaN (categorias que não tiveram vendas em alguma loja) com 0
vendas_por_categoria_loja_pivot = vendas_por_categoria_loja_pivot.fillna(0)

# Gerando o gráfico de barras agrupadas
ax = vendas_por_categoria_loja_pivot.plot(kind='bar', figsize=(15, 7))

plt.title('Comparativo de Vendas por Categoria do Produto entre as Lojas')
plt.xlabel('Categoria do Produto')
plt.ylabel('Vendas (R$)')
plt.xticks(rotation=90)
plt.legend(title='Loja')
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
# Calcula a média da coluna 'Avaliação' para cada loja
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média de Avaliação da Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de Avaliação da Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de Avaliação da Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de Avaliação da Loja 4: {media_avaliacao_loja4:.2f}")

# DataFrame para visualização
media_avaliacao_por_loja = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Média de Avaliação': [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]
})

print("\nMédia de Avaliação por Loja:")
print(media_avaliacao_por_loja)

# Gráfico de barras para comparar as médias de avaliação
media_avaliacao_por_loja.plot(kind='area', x='Loja', y='Média de Avaliação', legend=False, title='Média de Avaliação por Loja')
plt.ylabel('Média de Avaliação')
plt.xticks(rotation=0)
plt.ylim(0, 5) # Define o limite do eixo y para a escala de avaliação (0 a 5)
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Para cada loja, identificar o produto mais vendido e o menos vendido com base no preço
print("Produtos Mais e Menos Vendidos em cada Loja:")

for i, df in enumerate([loja, loja2, loja3, loja4]):
    loja_nome = f"Loja {i+1}"

    # Agrupar por nome do produto e somar os preços (vendas)
    vendas_por_produto = df.groupby('Produto')['Preço'].sum()

    # Identificar o produto com a maior venda
    produto_mais_vendido = vendas_por_produto.idxmax()
    venda_mais_vendida = vendas_por_produto.max()

    # Identificar o produto com a menor venda (ignorando produtos com venda 0 se houver)
    # Filtrar produtos com vendas maiores que 0 antes de encontrar o mínimo
    vendas_positivas = vendas_por_produto[vendas_por_produto > 0]
    if not vendas_positivas.empty:
        produto_menos_vendido = vendas_positivas.idxmin()
        venda_menos_vendida = vendas_positivas.min()
    else:
        produto_menos_vendido = "N/A (nenhum produto com venda positiva)"
        venda_menos_vendida = 0

    print(f"\n{loja_nome}:")
    print(f"  Produto Mais Vendido: '{produto_mais_vendido}' (Venda: R${venda_mais_vendida:,.2f})")
    print(f"  Produto Menos Vendido: '{produto_menos_vendido}' (Venda: R${venda_menos_vendida:,.2f})")

# Opcional: Visualizar os top 10 produtos mais vendidos e menos vendidos para cada loja
def plot_top_bottom_products(df, loja_nome, top_n=10):
    vendas_por_produto = df.groupby('Produto')['Preço'].sum()

    # Top N mais vendidos
    top_sellers = vendas_por_produto.nlargest(top_n)
    top_sellers.plot(kind='bar', figsize=(10, 5), title=f'Top {top_n} Produtos Mais Vendidos - {loja_nome}')
    plt.ylabel('Vendas (R$)')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

    # Top N menos vendidos (entre aqueles com venda > 0)
    vendas_positivas = vendas_por_produto[vendas_por_produto > 0]
    if not vendas_positivas.empty:
        bottom_sellers = vendas_positivas.nsmallest(top_n)
        bottom_sellers.plot(kind='bar', figsize=(10, 5), title=f'Top {top_n} Produtos Menos Vendidos - {loja_nome}')
        plt.ylabel('Vendas (R$)')
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
    else:
        print(f"Não há produtos com vendas positivas para listar os menos vendidos na {loja_nome}.")


plot_top_bottom_products(loja, 'Loja 1')
plot_top_bottom_products(loja2, 'Loja 2')
plot_top_bottom_products(loja3, 'Loja 3')
plot_top_bottom_products(loja4, 'Loja 4')
```

# 5. Frete Médio por Loja


```python
# Calcula a média da coluna 'Frete' para cada loja
media_frete_loja1 = loja['Frete'].mean()
media_frete_loja2 = loja2['Frete'].mean()
media_frete_loja3 = loja3['Frete'].mean()
media_frete_loja4 = loja4['Frete'].mean()

print(f"Média de Frete da Loja 1: R${media_frete_loja1:.2f}")
print(f"Média de Frete da Loja 2: R${media_frete_loja2:.2f}")
print(f"Média de Frete da Loja 3: R${media_frete_loja3:.2f}")
print(f"Média de Frete da Loja 4: R${media_frete_loja4:.2f}")

# DataFrame para visualização
media_frete_por_loja = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Média de Frete': [media_frete_loja1, media_frete_loja2, media_frete_loja3, media_frete_loja4]
})

print("\nMédia de Frete por Loja:")
print(media_frete_por_loja)

# Gráfico de pizza para comparar a proporção do frete médio por loja
plt.figure(figsize=(8, 8))
plt.pie(media_frete_por_loja['Média de Frete'], labels=media_frete_por_loja['Loja'], autopct='%1.1f%%', startangle=90)
plt.title('Distribuição do Frete Médio por Loja')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
